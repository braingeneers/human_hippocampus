```python
import numpy as np
import matplotlib.pyplot as plt

# Set the seed for reproducibility
np.random.seed(42)

# Generate data for Neuron A to B with mean of -6ms and std deviation of 4ms
latencies_ab_tighter = np.random.normal(loc=-6, scale=4, size=500)

# Generate bimodal data for Neuron B to C
latencies_bc_bimodal1 = np.random.normal(loc=-10, scale=3, size=250)  # First peak
latencies_bc_bimodal2 = np.random.normal(loc=5, scale=3, size=250)   # Second peak
latencies_bc_bimodal = np.concatenate((latencies_bc_bimodal1, latencies_bc_bimodal2))

fig, axs = plt.subplots(1, 2, figsize=(20, 6))

# Plot for Neuron A to B
n, bins, patches = axs[0].hist(latencies_ab_tighter, bins=20, alpha=0.75)
axs[0].axvline(latencies_ab_tighter.mean(), color='r', linestyle='dashed', linewidth=2, label=f'Mean: {latencies_ab_tighter.mean():.2f}ms')

# Calculate FWHM
half_max = max(n) / 2
bin_centers = 0.5*(bins[1:] + bins[:-1])
lower_bin = bin_centers[n > half_max].min()
upper_bin = bin_centers[n > half_max].max()
fwhm = upper_bin - lower_bin

axs[0].axvline(latencies_ab_tighter.mean() - fwhm / 2, color='g', linestyle='dashed', linewidth=2, label=f'FWHM: {fwhm:.2f}ms')
axs[0].axvline(latencies_ab_tighter.mean() + fwhm / 2, color='g', linestyle='dashed', linewidth=2)

axs[0].set(title="Unimodal Spike Time Latency Distribution from Neuron A to B", xlabel="Latency (ms)", ylabel="Frequency")
axs[0].legend()
axs[0].grid(True)

# Plot for Neuron B to C
axs[1].hist(latencies_bc_bimodal, bins=20, alpha=0.75, color='orange')
axs[1].axvline(latencies_bc_bimodal.mean(), color='r', linestyle='dashed', linewidth=2, label=f'Mean: {latencies_bc_bimodal.mean():.2f}ms')

axs[1].set(title="Bimodal Spike Time Latency Distribution from Neuron B to C", xlabel="Latency (ms)", ylabel="Frequency")
axs[1].legend()
axs[1].grid(True)

# Adjust xlim for both subplots to ensure they cover the full range of data
for ax in axs:
    latencies = latencies_ab_tighter if ax == axs[0] else latencies_bc_bimodal
    ax.set_xlim(min(latencies.min(), -latencies.max()), max(-latencies.min(), latencies.max()))

plt.tight_layout()
plt.show()
```

```python
import numpy as np
import matplotlib.pyplot as plt

# Set the seed for reproducibility
np.random.seed(42)

# Generate data for Neuron A to B and B to C
latencies_ab_tighter = np.random.normal(loc=-6, scale=4, size=500)
latencies_bc_bimodal1 = np.random.normal(loc=-10, scale=3, size=250)
latencies_bc_bimodal2 = np.random.normal(loc=5, scale=3, size=250)
latencies_bc_bimodal = np.concatenate((latencies_bc_bimodal1, latencies_bc_bimodal2))

fig, axs = plt.subplots(1, 2, figsize=(16, 5))  # Adjusted dimensions for compactness

# Customizing font sizes
title_fontsize = 14
label_fontsize = 12
legend_fontsize = 10
tick_fontsize = 10

# Plot for Neuron A to B
n, bins, patches = axs[0].hist(latencies_ab_tighter, bins=20, alpha=0.75)
mean_val = latencies_ab_tighter.mean()
axs[0].axvline(mean_val, color='r', linestyle='dashed', linewidth=2, label=f'Mean: {mean_val:.2f}ms')

# Calculate and plot FWHM
half_max = max(n) / 2
bin_centers = 0.5*(bins[1:] + bins[:-1])
lower_bin = bin_centers[n > half_max].min()
upper_bin = bin_centers[n > half_max].max()
fwhm = upper_bin - lower_bin

axs[0].axvline(mean_val - fwhm / 2, color='g', linestyle='dashed', linewidth=2, label=f'FWHM: {fwhm:.2f}ms')
axs[0].axvline(mean_val + fwhm / 2, color='g', linestyle='dashed', linewidth=2)

axs[0].set(title="Unimodal Spike Time Latency Distribution from Neuron A to B",
           xlabel="Latency (ms)", ylabel="Frequency")
axs[0].title.set_fontsize(title_fontsize)
axs[0].xaxis.label.set_fontsize(label_fontsize)
axs[0].yaxis.label.set_fontsize(label_fontsize)
axs[0].tick_params(axis='both', which='major', labelsize=tick_fontsize)
axs[0].legend(fontsize=legend_fontsize)
axs[0].grid(True)

# Plot for Neuron B to C
axs[1].hist(latencies_bc_bimodal, bins=20, alpha=0.75, color='orange')
mean_val_bc = latencies_bc_bimodal.mean()
axs[1].axvline(mean_val_bc, color='r', linestyle='dashed', linewidth=2, label=f'Mean: {mean_val_bc:.2f}ms')

axs[1].set(title="Bimodal Spike Time Latency Distribution from Neuron B to C",
           xlabel="Latency (ms)", ylabel="Frequency")
axs[1].title.set_fontsize(title_fontsize)
axs[1].xaxis.label.set_fontsize(label_fontsize)
axs[1].yaxis.label.set_fontsize(label_fontsize)
axs[1].tick_params(axis='both', which='major', labelsize=tick_fontsize)
axs[1].legend(fontsize=legend_fontsize)
axs[1].grid(True)

# Adjust xlim for both subplots to cover the full range of data and ensure readability
for ax in axs:
    latencies = latencies_ab_tighter if ax == axs[0] else latencies_bc_bimodal
    ax.set_xlim(min(latencies.min(), -latencies.max()), max(-latencies.min(), latencies.max()))

plt.tight_layout()
plt.show
```
