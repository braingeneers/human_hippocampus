## Polar Vectors

```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
spatial_range = 0, 100
bins = 6

distance_bins = np.linspace(spatial_range[0], spatial_range[1], bins+1)
angle_bins = np.linspace(0, 2*np.pi, bins+1, endpoint=True)
```

```python
# Generating 5 random vectors with the specified seed for reproducibility
np.random.seed(3)  # Using the same seed as before
angles = np.random.uniform(0, 2 * np.pi, 5)  # Random angles
lengths = np.random.uniform(0.5, 1, 5) 

# Convert polar coordinates (angles and lengths) to Cartesian coordinates (x, y)
x_coords = lengths * np.cos(angles)
y_coords = lengths * np.sin(angles)

# Calculate the average x and y components
avg_x = np.mean(x_coords)
avg_y = np.mean(y_coords)

# Convert the average Cartesian coordinates back to polar coordinates
avg_angle_corrected = np.arctan2(avg_y, avg_x)
if avg_angle_corrected < 0:
    avg_angle_corrected += 2 * np.pi  # Adjusting the angle to be in the range [0, 2π]
avg_length_corrected = np.sqrt(avg_x**2 + avg_y**2)

# Setting up the plot again with the corrected average vector
fig, ax = plt.subplots(figsize=(2.5, 2.5), subplot_kw={'projection': 'polar'})

# Plotting the individual vectors
for angle, length in zip(angles, lengths):
    ax.plot([0, angle], [0, length], marker='o', color='red', linewidth=1.5, markersize=4)

# Plotting the corrected average vector
ax.plot([0, avg_angle_corrected], [0, avg_length_corrected], marker='o', color='black', linewidth=3, markersize=6, label=' Average Vector')

# Adjusting plot settings for clarity
ax.legend(loc='upper center', fontsize='small', framealpha=0.5)
ax.set_ylim(0, 1)
ax.tick_params(axis='both', which='major', labelsize=8)
plt.show(), (avg_angle_corrected, avg_length_corrected)
```

## Spatial Range

```python
def spatial_range_plot(dim=64, min_range=4, max_range=24):
    # Create a grid of neuron positions
    neuron_x, neuron_y = np.meshgrid(range(dim), range(dim))
    plt.figure(figsize=(6, 6))
    
    # Calculate the spatial range from the red cross at (56, 8)
    distance_from_red_cross = np.sqrt((neuron_x - 56)**2 + (neuron_y - 8)**2)
    
    # Define color for each neuron based on the spatial range condition
    colors = np.where((distance_from_red_cross > min_range) & (distance_from_red_cross < max_range), 'palevioletred', '#add8e6')
    
    # Plotting neurons, color depends on their spatial range from the red cross
    for i in range(dim):
        for j in range(dim):
            plt.scatter(j, i, s=3, c=colors[i, j], marker='s', alpha=0.6)
    
    # Plotting the red cross at (56, 8)
    plt.scatter(56, 8, s=200, c='red', marker='x', label='Location of Interest')
    
    # Adding legend for the different elements
    plt.scatter([], [], s=10, c='palevioletred', marker='s', label='Locations in Range')
    plt.scatter([], [], s=10, c='#add8e6', marker='s', label='Electrodes')
    plt.legend(loc='lower left')
    plt.xlabel('X Position')
    plt.ylabel('Y Position')
    plt.gca().invert_yaxis()
    plt.show()
```

```python
spatial_range_plot(dim=64, min_range=8, max_range=56)
```

```python
def spatial_range_plot_with_arrow(dim=64, min_range=2, max_range=26):
    # Create a grid of neuron positions
    neuron_x, neuron_y = np.meshgrid(range(dim), range(dim))
    plt.figure(figsize=(6, 6))

    # Calculate the spatial range from the red cross at (56, 8)
    distance_from_red_cross = np.sqrt((neuron_x - 56)**2 + (neuron_y - 8)**2)
    
    # Define color for each neuron based on the spatial range condition
    colors = np.where((distance_from_red_cross > min_range) & (distance_from_red_cross < max_range), 'palevioletred', '#add8e6')

    # Initialize variables to compute average position of red shaded electrodes
    sum_x, sum_y, count = 0, 0, 0

    # Plotting neurons, color depends on their spatial range from the red cross
    for i in range(dim):
        for j in range(dim):
            if colors[i, j] == 'palevioletred':
                # Accumulate positions of red shaded electrodes
                sum_x += j
                sum_y += i
                count += 1
            plt.scatter(j, i, s=3, c=colors[i, j], marker='s', alpha=0.6)

    # Calculate average position of red shaded electrodes
    if count > 0:  # Prevent division by zero
        avg_x, avg_y = sum_x / count, sum_y / count
        # Plotting the arrow from the red cross to the average position of red shaded electrodes
        plt.arrow(56, 8, avg_x - 56, avg_y - 8, head_width=1.5, head_length=2, fc='black', ec='black')
    
    # Plotting the red cross at (56, 8)
    plt.scatter(56, 8, s=200, c='red', marker='x', label='Location of Interest')

    # Adding legend for the different elements
    plt.scatter([], [], s=10, c='palevioletred', marker='s', label='Locations in Range')
    plt.scatter([], [], s=10, c='#add8e6', marker='s', label='Electrodes')
    plt.legend(loc='lower left')
    plt.xlabel('X Position')
    plt.ylabel('Y Position')
    plt.gca().invert_yaxis()
    plt.show()
```

```python
spatial_range_plot_with_arrow(dim=64, min_range=2, max_range=26)
```

```python
def spatial_range_plot_with_vector(dim=64, min_range=2, max_range=26):
    # Create a grid of neuron positions
    neuron_x, neuron_y = np.meshgrid(range(dim), range(dim))
    plt.figure(figsize=(6, 6))

    # Calculate the spatial range from the red cross at (56, 8)
    distance_from_red_cross = np.sqrt((neuron_x - 56)**2 + (neuron_y - 8)**2)
    
    # Define color for each neuron based on the spatial range condition
    colors = np.where((distance_from_red_cross > min_range) & (distance_from_red_cross < max_range), 'palevioletred', '#add8e6')

    # Initialize variables to accumulate vectors
    vector_sum_x, vector_sum_y, count = 0, 0, 0

    # Plotting neurons, color depends on their spatial range from the red cross
    for i in range(dim):
        for j in range(dim):
            if colors[i, j] == 'palevioletred':
                # Accumulate vector components from the red cross to the electrode
                vector_sum_x += (j - 56)
                vector_sum_y += (i - 8)
                count += 1
            plt.scatter(j, i, s=3, c=colors[i, j], marker='s', alpha=0.6)

    # Plotting the red cross at (56, 8)
    plt.scatter(56, 8, s=200, c='red', marker='x', label='Location of Interest')

    # Calculate and plot the average vector if there are red shaded electrodes
    if count > 0:
        avg_vector_x = vector_sum_x / count
        avg_vector_y = vector_sum_y / count
        # Print positions of tail and head of the average vector
        print(f"Arrow's Tail (Red Cross Position): (56, 8)")
        print(f"Arrow's Head (Relative Position): ({56 + avg_vector_x}, {8 + avg_vector_y})")

        # Plotting the average vector as an arrow
        plt.arrow(56, 8, avg_vector_x, avg_vector_y, head_width=1.5, head_length=2, fc='black', ec='black')
    
    # Adding legend and other plot configurations
    plt.scatter([], [], s=10, c='palevioletred', marker='s', label='Locations in Range')
    plt.scatter([], [], s=10, c='#add8e6', marker='s', label='Electrodes')
    plt.legend(loc='lower left')
    plt.xlabel('X Position')
    plt.ylabel('Y Position')
    plt.gca().invert_yaxis()
    plt.show()
```
