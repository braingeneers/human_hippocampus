```python
import h5py

def process_group(group):
    if isinstance(group, h5py.Group):
        subgroup_names = list(group.keys())
        subgroup_data = {}
        for subgroup_name in subgroup_names:
            subgroup = group[subgroup_name]
            subgroup_data[subgroup_name] = process_group(subgroup)
        return subgroup_data
    else:
        return group[()]  # Return dataset as is
    
def read_h5_data(file_path):
    data = {}
    with h5py.File(file_path, 'r') as file:
        group_names = list(file.keys())
        for group_name in group_names:
            group = file[group_name]
            if isinstance(group, h5py.Group):  # Check if group is an instance of h5py.Group
                subgroup_names = list(group.keys())
                print(f"Group: {group_name}")
                print(f"Subgroups: {subgroup_names}")
            data[group_name] = process_group(group)
    return data

file_path = '/home/jovyan/work/Human_Hippocampus/data/ephys/2023-04-02-hc328_rec/original/data/2023_04_02_hc328_0.raw.h5'
result_data = read_h5_data(file_path)
print("********************")
print(result_data)
```
